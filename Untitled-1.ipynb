```python
import pandas as pd

df = pd.read_excel('D:/downloads/ai-queries.xlsx')
print(df.shape)
data = {q: (q, *vs) for _, q, *vs in df.itertuples()}
df = pd.DataFrame(
    [
        {
            'query': q,
            'target': tgt,
            'pos': pos,
            'neg': neg,
            'link': ln,
        }
        for q, ln, tgt, pos, neg in data.values()
    ]
)
print(df.shape)
df.to_excel('D:/downloads/ai-queries-fixed.xlsx')
df
```
